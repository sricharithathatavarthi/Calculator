```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("Titanic-Dataset.csv")

# Select numeric columns
numeric_cols = ['Age','Fare','SibSp','Parch','Pclass','Survived']

# Compute mean, median, mode, std, min, max for each numeric column
summary_stats = pd.DataFrame({
    'Mean': df[numeric_cols].mean(),
    'Median': df[numeric_cols].median(),
    'Mode': df[numeric_cols].mode().iloc[0],
    'Std Dev': df[numeric_cols].std(),
    'Min': df[numeric_cols].min(),
    'Max': df[numeric_cols].max()
})

summary_stats
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for Age
plt.figure(figsize=(6,4))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title("Age Distribution")
plt.show()

# Boxplot for Fare
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Fare'])
plt.title("Fare Boxplot")
plt.show()

```

```python
# Pairplot for selected numeric features
sns.pairplot(df[['Survived','Age','Fare','Pclass']], hue='Survived')
plt.show()

# Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df[['Survived','Age','Fare','Pclass','SibSp','Parch']].corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
# Survival rate by class
sns.barplot(x="Pclass", y="Survived", data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()

# Survival rate by gender
sns.barplot(x="Sex", y="Survived", data=df)
plt.title("Survival Rate by Gender")
plt.show()

# Check outliers in Fare
sns.boxplot(x="Survived", y="Fare", data=df)
plt.title("Fare vs Survival (Outliers visible)")
plt.show()
```
